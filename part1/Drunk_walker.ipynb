#Diffusion in 1 dimension: the drunk walker

```
#libraries
import random as r 
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss
```

```
def position(initial_pos,p,nbsteps): # calculate final position after N steps 
  pos=initial_pos # initial position of drunk walker
  for i in range(nbsteps): 
    a=r.random() # takes random number between 0 and 1 
    if p>a:
      pos=pos+1 # move one step to forward 
    else:
      pos=pos-1 # move one step to backward
  return pos

def list_positions(initial_pos,p,nbsteps,nbtimes):    # return positions x times as list
  list_pos=[]
  for i in range(nbtimes): 
    list_pos.append(position(initial_pos,p,nbsteps))
  return list_pos  

def histogram(initial_pos,p,nbsteps,nbtimes):   # plot histogram
  plt.title(f'Calculate positions {nbtimes} times for {nbsteps} steps')
  plt.xlabel('Position')
  plt.ylabel('Probability density')
  plt.hist(list_positions(initial_pos,p,nbsteps,nbtimes),bins=20,density=True,edgecolor='black',color='salmon');

def Gaussian(initial_pos,p,nbsteps,nbtimes): 
  lpos=list_positions(initial_pos,p,nbsteps,nbtimes)
  std=np.std(lpos)  # calculate standard deviation
  mean=np.mean(lpos) # calculate mean
  print(f'sigma {nbsteps} is: {std}\nmean is: {mean}')
  
  histogram(initial_pos,p,nbsteps,nbtimes)
  xmin, xmax = plt.xlim()
  x = np.linspace(xmin, xmax, 1000)
  pdf = ss.norm.pdf(x, mean, std) # normal probability density function
  plt.plot(x, pdf,'blue',linewidth=3) # plot curve to histogram
  plt.show()
```

```
init=0
p=0.5
N=1000

print(f'The final position for probability {p} and for {N} steps is: {position(init,p,N)}')
```

```
nbtimes=100

# histogram(init,p,N,nbtimes)
Gaussian(init,p,N,nbtimes) 
```

```
steps=[20,50,100,200]
for i in steps:
  Gaussian(init,p,i,nbtimes)
```

```
p2=0.75

print(f'The final position for probability {p2} and for {N} steps is: {position(init,p2,N)}')
```
